```python
# =========================
# VISUALISASI DATA KUESIONER
# =========================

import pandas as pd
import matplotlib.pyplot as plt

# =========================
# LOAD DATA
# =========================
file_path = "data_kuesioner.xlsx"
df = pd.read_excel(file_path)

# Hapus kolom pertama jika itu ID / Partisipan
df = df.iloc[:, 1:]

# Ubah ke string agar tidak error tipe data
df_str = df.astype(str)

# =========================
# 1. BAR CHART
# Distribusi Jawaban Keseluruhan
# =========================
overall_counts = df_str.stack().value_counts()
```

```python
# =========================
# 1. BAR CHART
# Distribusi Jawaban Keseluruhan
# =========================
overall_counts = df_str.stack().value_counts()

plt.figure(figsize=(8,5))
overall_counts.plot(kind='bar')
plt.title("Distribusi Jawaban Keseluruhan")
plt.xlabel("Kategori Jawaban")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# =========================
# 2. PIE CHART
# Proporsi Jawaban Keseluruhan
# =========================
plt.figure(figsize=(6,6))
overall_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Proporsi Jawaban Keseluruhan")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
